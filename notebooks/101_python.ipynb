### 1.5 Functions (print, hash, max, open, etc. are [built-in functions](https://docs.python.org/3/library/functions.html) )

Fibonaci sequence
> $ u_{n+2} = u_{n+1} + u_{n},$ with $u_0 = 0$ and $u_1 = 1$ 

```python
def fib(n):
    a, b = 0, 1
    while a < n:
        print(a)
        a, b = b, a+b 
fib(100)
```

### 1.6 [Classes](https://docs.python.org/3.6/tutorial/classes.html)

```python
class Letter:
    def __init__(self, to, obj, message):
        self.to = to
        self.obj = obj
        self.message = message
        self.status = 'draft'
    
    def lenght(self):
        return len(self.message) 
    
    def send(self):
        if(self.lenght() > 1000):
            print('Can not send message with more that 1000 characters.')
        else:
            print('Message sent.')
            self.status = 'sent'
```

```python
new = Letter(to='email@my.self', obj='object', message='Hello!')
print(new)
print(new.to)
new.send()
```

## 5. Examples

### 5.1 Least-square solution $Ax=b$

```python
import numpy as np
import matplotlib.pyplot as plt
# Fit a line, ``y = mx + c``, through some noisy data-points:

x = np.array([0, 1, 2, 3])
y = np.array([-1, 0.2, 0.9, 2.1])

A = np.vstack([x, np.ones(len(x))]).T # vertical stack followed by a transposition.
print(A) 
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print(m, c)
```

```python
plt.plot(x, y, 'o', label='Original data', markersize=10)
plt.plot(x, m*x + c, 'r', label='Fitted line')
plt.legend()
```

### 5.2 Image patch extraction

```python
from skimage import io, transform 
import matplotlib.pyplot as plt
image = transform.resize(io.imread('../images/image3.jpg', 1), (300,300))
print('image shape: {}'.format(image.shape))
plt.imshow(image, cmap=plt.cm.gray)
patch_shape = (32,32)
```

```python
plt.imshow(image[100:132, 100:132])
```

```python
padded_im = np.pad(image, ((16,16), (16,16)), mode='constant')
print(padded_im.shape)
```

```python
plt.imshow(padded_im, cmap=plt.cm.gray)
```

```python
ids = np.zeros_like(image)
ids[::10, ::10] = 1
```

```python
plt.matshow(ids)
```

```python
x_ids, y_ids = np.nonzero(ids)
rand = np.random.permutation(range(len(x_ids)))
x_ids = x_ids[rand]
y_ids = y_ids[rand]
```

```python
k = 1
fig = plt.figure(figsize=(15,15))
for x, y in zip(x_ids[:100], y_ids[:100]):
    plt.subplot(10,10,k)
    k += 1
    plt.imshow(padded_im[x:x+patch_shape[0], y:y+patch_shape[1]], cmap=plt.cm.gray)
```
