## 1 Sparse Coding

Given a dictionary $D \in \mathbb{R}^{d\times p} $ and a dataset $\{\mathbf{x}^{(t)} \in \mathbb{R}^d\}_{t=1}^T$, compute

$ \underset{\mathbf{h} \in  \mathbb{R}^p}{\text{arg}\min } \ \frac{1}{T} \sum_{t=1}^T \|\mathbf{x}^{(t)} - \mathbf{Dh}^{(t)}\|_2^2 + 
\alpha \|\mathbf{h}^{(t)}\|_1$


### 1.1- Iterative Shrinkage and Thresholding Algorithm (ISTA)

```python
def ISTA(X, D, alpha, learning_rate=.5, maxiter=100, eps=1e-3):
    H = np.zeros((D.shape[1], X.shape[1]))
    it = 0
    error = np.inf
    cmp_err = lambda h: np.linalg.norm( X - np.matmul(D, h))**2 + alpha*np.abs(h).sum()
    shrink  = lambda a, b:  np.sign(a)*np.maximum(np.abs(a) - b, 0)
    while error > eps and it < maxiter: 
        H = H + learning_rate*np.matmul(D.T, X - np.matmul(D, H))
        H = shrink(H, np.ones(H.shape)*learning_rate*alpha)
        error = cmp_err(H)
        it += 1
    return H
```

### 1.2- Coordinate Descent

```python
def CD(X, D, alpha, maxiter=100, eps=1e-3):
    H = np.zeros((D.shape[1], X.shape[1]))
    it = 0
    error = np.inf
    cmp_err = lambda h: np.linalg.norm( X - np.matmul(D, h))**2 + alpha*np.abs(h).sum()
    shrink  = lambda a, b:  np.sign(a)*np.maximum(np.abs(a) - b, 0)
    while error > eps and it < maxiter:
        for j in range(H.shape[0]):
            H[j,:] = H[j,:] + np.matmul(D[:,j].T, X - np.matmul(D, H))/np.linalg.norm(D[:,j])
            H[j,:] = shrink(H[j,:], np.ones(H[j,:].shape)*alpha)
        error = cmp_err(H)
        it += 1
    return H
```

## 2 Dictionary Learning

Given a dataset $\{\mathbf{x}^{(t)} \in \mathbb{R}^d\}_{t=1}^T$, compute

$ \underset{\mathbf{h} \in  \mathbb{R}^p, D \in \mathbb{R}^{d\times p} }{\text{arg}\min } \ \frac{1}{T}
\sum_{t=1}^T \|\mathbf{x}^{(t)} - \mathbf{Dh}^{(t)}\|_2^2 + 
\alpha \|\mathbf{h}^{(t)}\|_1$


### 2.1- Block-Coordinate Descent

Based on the 2009 paper of  Y. Li and  S. Osher, 
[Coordinate descent optimization for $\ell_1$ minimization with application to compressed sensing; a greedy algorithm](http://www.aimsciences.org/article/doi/10.3934/ipi.2009.3.487), Inverse Problems and Imaging.

```python
def BCD(X, H, maxiter=100, eps=1e-3, n_atoms=100, init_D=None, B=None, A=None):
    D = init_D
    if D is None:
        D = np.random.randn(X.shape[0], n_atoms)
        D = np.array([D[:,j]/(np.linalg.norm(D[:,j]) + 1e-5) for j in range(D.shape[1])]).T # unit l2-norm
    B = B if B is not None else np.matmul(X, H.T)
    A = A if A is not None else np.matmul(H,H.T) 
    it = 0
    error = np.inf
    cmp_err = lambda D: np.linalg.norm( X - np.matmul(D, H))**2
    while error > eps and it < maxiter:
        for j in range(D.shape[1]):
            D[:,j] = (1./max(A[j,j], 1e-5)) *( B[:,j] - np.matmul(D, A[:,j])) + D[:,j]
            D[:,j] = D[:,j]/(np.linalg.norm(D[:,j]) + 1e-5)
        error = cmp_err(D)
        it += 1
    return D

def DicoLearn_BCD(X, alpha, learning_rate=.5, maxiter=100, dico_iter=100, sc_iter=10,
                  eps=1e-3, n_atoms=100, init_D=None, verbose=0, sc_algo='ista'):
    D = init_D
    if D is None:
        D = np.random.randn(X.shape[0], n_atoms)
        D = np.array([D[:,j]/(np.linalg.norm(D[:,j]) + 1e-5) for j in range(D.shape[1])]).T # unit l2-norm
    for k in range(maxiter):
        if sc_algo == 'ista':
            H = ISTA(X, D, alpha, learning_rate, sc_iter, eps)
        else:
            H = CD(X, D, alpha, sc_iter, eps)
        D = BCD(X, H, dico_iter, eps, n_atoms, D)
        if verbose:
            print '[DicoLearn_BCD] iteration %d' %k
    return D, H
```

### 2.2- Online Dictionary Learning

Based on the ICML'09 paper of J. Mairal, F. Bach, J. Ponce, and G. Sapiro, 
[Online Dictionary Learning for Sparse Coding](https://www.di.ens.fr/~fbach/mairal_icml09.pdf).

```python
def DicoLearn_ODL(X, alpha, batch_size=1, learning_rate=.5, maxiter=100, dico_iter=10, sc_iter=10,
                  eps=1e-3, n_atoms=100, init_D=None, beta=.5, verbose=0, sc_algo='ista'):
    D = init_D
    if D is None:
        D = np.random.randn(X.shape[0], n_atoms)
        D = np.array([D[:,j]/(np.linalg.norm(D[:,j]) + 1e-5) for j in range(D.shape[1])]).T # unit l2-norm
    B = 0; A = 0
    for k in range(maxiter):
        ids = np.random.permutation(np.arange(X.shape[1]))
        for t in range(0, len(ids)/batch_size, batch_size):   
            if sc_algo == 'ista': 
                h = ISTA(X[:,t:t+batch_size], D, alpha, learning_rate, sc_iter, eps)
            else:
                h = CD(X[:,t:t+batch_size], D, alpha, sc_iter, eps)
            B = beta*B + np.matmul(X[:,t:t+batch_size], h.T)
            A = beta*A + np.matmul(h, h.T)
            D = BCD(X[:,t:t+batch_size], h, maxiter=dico_iter, eps=eps, n_atoms=n_atoms, init_D=D, )
        if verbose:
            print '[DicoLearn_OLD] iteration %d' %k
    return D
```

## 3 Application

### 3.1- Image reading and Patch extraction

```python
image = transform.resize(io.imread('../images/image3.jpg', 1), (300,300))
patch_shape = (32,32)
plt.imshow(image, cmap=plt.cm.gray);plt.show()
```

```python
patches = extract_patches(image, patch_shape=patch_shape, extraction_step=20)
patches = patches.reshape((-1,)+ patch_shape)
patches = patches.reshape(patches.shape[0], -1)
patches = np.random.permutation(patches).T
#plot_patches(patches.T[:25].reshape((25,)+patch_shape), 26)
```

```python
def plot_patches(X, n=101):
    plt.figure(figsize=(20,20))
    for i in range(1, n):
        plt.subplot(np.sqrt(n-1),np.sqrt(n-1),i)
        plt.imshow(X[i-1], cmap=plt.cm.gray)
```

### 3.2- Applying the Block-coordinate Dictionary Learning Algorithm 

```python
D, H = DicoLearn_BCD(patches, alpha=1., learning_rate=.01,
                     maxiter=10, dico_iter=10, sc_iter=10, n_atoms=100, verbose=0, sc_algo='CD')
```

```python
plt.bar(range(100),H[:,10]);plt.show()
```

```python
plot_patches(D.T.reshape((100,) + patch_shape))
```

### 3.2- Applying the Online Dictionary Learning Algorithm

```python
D_odl = DicoLearn_ODL(patches, alpha=1., batch_size=3,
                     learning_rate=.01, maxiter=10, dico_iter=10, sc_iter=10, n_atoms=100,
                     sc_algo='CD')
```

```python
plot_patches(D_odl.T.reshape((100,) + patch_shape))
```
